### In order to optimize memory we will convert data types

```python
def optimize_memory(dataset):
    for column in dataset.columns:
        if dataset[column].dtype == np.int64:
            maxVal = dataset[column].max()
            if maxVal < 120:
                dataset[column] = dataset[column].astype(np.int8)
            elif maxVal < 32767:
                dataset[column] = dataset[column].astype(np.int16)
            else:
                dataset[column] = dataset[column].astype(np.int32)
            
        if dataset[column].dtype == np.float64:
            maxVal = dataset[column].max()
            minVal = dataset[dataset[column]>0][column]
            if maxVal < 120 and minVal>0.01 :
                dataset[column] = dataset[column].astype(np.float16)
            else:
                dataset[column] = dataset[column].astype(np.float32)
```

# K-Means Clustering Analysis

This section presents the application of K-Means clustering to analyze our dataset. K-Means clustering is a method of vector quantization that is popular for cluster analysis in data mining. K-Means clustering aims to partition n observations into k clusters in which each observation belongs to the cluster with the nearest mean (cluster centers or cluster centroid), serving as a prototype of the cluster.

## Objective

The objective of this analysis is to explore how the number of clusters (k) affects the grouping of data points and to identify the optimal number of clusters based on the silhouette score. The silhouette score is a measure of how similar an object is to its own cluster compared to other clusters.

## Data Description

The dataset contains features extracicids2017s briefly]. Due to the extensive size of the dataset, the following analysis will only use a subset to ensure reasonable computation times for demonstration purposes. Full dataset analysis may require significant computation time and resources.



```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

def kmeans_analysis(data, k_range):
    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=0)
        cluster_labels = kmeans.fit_predict(data)

        # Silhouette score for cluster evaluation
        silhouette_avg = silhouette_score(data, cluster_labels)
        print(f"Number of clusters: {k}, Silhouette Score: {silhouette_avg:.4f}")

        # Plotting histogram of cluster labels
        plt.figure(figsize=(8, 4))
        plt.hist(cluster_labels, bins=np.arange(k+1)-0.5, rwidth=0.8, color='blue', alpha=0.7)
        plt.title(f'Histogram of K-Means Cluster Labels (k={k})')
        plt.xlabel('Cluster Label')
        plt.ylabel('Frequency')
        plt.show()

# Example usage
# Assuming 'data' is your DataFrame that has been preprocessed and ready for analysis
k_range = range(2, 5)  # Example range, can be adjusted based on your requirement
# Replace 'data' with your actual DataFrame
# kmeans_analysis(data, k_range)
```

```python
# traning the model

# Perform K-Means clustering analysis with different k values
kmeans_analysis(sample_df, k_range)
```

# Local Outlier Factor (LOF) Analysis

This section is dedicated to identifying outliers within our dataset using the Local Outlier Factor (LOF) method. LOF is an algorithm designed to detect outliers by measuring the local deviation of a given data point with respect to its neighbors.

## Objective

The objective of this analysis is to detect anomalies or outliers in our dataset by considering the density of their local neighborhoods. We analyze how varying the number of neighbors influences the identification of outliers.

## Data Descrour data]. We focus on a subset of data for computational efficiency, noting that extending this analysis to the entire dataset would significantly increase the computational demands.



```python
import numpy as np
import pandas as pd
from sklearn.neighbors import LocalOutlierFactor
import matplotlib.pyplot as plt

# Function to apply Local Outlier Factor (LOF) and evaluate with anomaly scores
def lof_analysis(data, n_neighbors_range):
    for n_neighbors in n_neighbors_range:
        lof = LocalOutlierFactor(n_neighbors=n_neighbors)
        lof.fit(data)
        # LOF does not have 'labels_', but we can use negative_outlier_factor_ to get the anomaly score
        lof_scores = -lof.negative_outlier_factor_  # Invert scores for easier interpretation

        # Calculate the threshold for outliers
        threshold = np.percentile(lof_scores, 95)  # Adjust the percentile to control strictness
        num_outliers = np.sum(lof_scores > threshold)

        # Print the analysis results
        print(f"Number of neighbors: {n_neighbors}, Number of outliers: {num_outliers}")

        # Plot histogram of LOF scores
        plt.hist(lof_scores, bins=50, color='blue', alpha=0.7)
        plt.title(f'Histogram of LOF Scores (n_neighbors={n_neighbors})')
        plt.xlabel('LOF Score')
        plt.ylabel('Frequency')
        plt.axvline(x=threshold, color='red', linestyle='--', label='Threshold for Outliers')
        plt.legend()
        plt.show()

# Example usage
# Assuming 'sample_df' is your DataFrame that has been preprocessed and ready for analysis
n_neighbors_range = range(20, 50, 10)  # Example range, adjust based on your dataset size
# Ensure that `sample_df` is defined somewhere in your actual use
# lof_analysis(sample_df, n_neighbors_range)
```

```python

lof_analysis(sample_df, n_neighbors_range)
```

# Conclusion

The LOF analysis consistently identified 5 outliers across different settings for the number of neighbors (20, 30, 40). This consistency indicates robustness in our outlier detection process, regardless of the variation in neighborhood size.

**Note on Scalability**: Similar to other data-intensive analyses, increasing the dataset size or the range of neighbors considered will proportionally increase computation time. It is important to balance the depth of outlier detection with available computational resources, especially when working with very large datasets.

Future investigations could explore different thresholds for outlier determination or combine LOF with other anomaly detection techniques to enhance the robustness and accuracy of outlier detection in larger datasets.


# Comparison of LOF and K-Means Clustering

In this section, we compare two distinct methods used in data analysis: Local Outlier Factor (LOF) for outlier detection and K-Means clustering for data segmentation. Both methods have unique characteristics and are used for different purposes in data analysis.

## Objectives

- **LOF**: Identifies anomalies in the data by focusing on the density of local neighborhoods.
- **K-Means**: Partitions the data into clusters to minimize the variance within each cluster.

## Data Descs briefly]. This comparison will help illustrate the effectiveness of each method under similar conditions.
